```python
import pandas as pd
import csv

# Path to your data file
file_path = "datasample.txt"  # Or .txt if you saved as plain text

# Define column names
columns = [
    "Day Record", "Register Date", "Source", "MemberType", "Username", 
    "Fullname", "Line", "DOB", "VIP", "Phone", 
    "Email"
]

# Read the file
df = pd.read_csv(
    file_path,
    sep="\t",                  # tab-separated
    names=columns,             # no header row in file
    header=None,
    engine="python",           # robust with multi-line fields
    quotechar='"',             # honor quotes
    quoting=csv.QUOTE_MINIMAL, # allow newlines inside quoted Notes
    doublequote=True,
    na_values=["null"],
    dtype=str,                 # keep everything as string (incl. '@...')
    keep_default_na=False
)

df 
```
