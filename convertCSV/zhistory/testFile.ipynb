```python
import pandas as pd

# Path to your data file
file_path = "crmsample.txt"  # Or .txt if you saved as plain text

# Define column names
columns = [
    "Registration date", "Member Type", "Username", "Fullname", "Line", 
    "Phone", "Contact", "CallDate", "NumOfDays", "AccountStatus", 
    "CallStatus", "Status", "Amount", "CRM1", "CRM2", "Notes"
]

# Read the file
df = pd.read_csv(
    file_path,
    sep="\t",                # Tab-separated
    names=columns,           # Assign our column names
    header=None,             # No header row in file
    quoting=3,               # Keep quotes as-is for multi-line text
    na_values=["null"],      # Convert "null" to NaN
    engine="python"          # More robust parsing for multi-line text
)

# Move Username column to the first position
df = df[["Username"] + [col for col in df.columns if col != "Username"]] 
df.head()
```
