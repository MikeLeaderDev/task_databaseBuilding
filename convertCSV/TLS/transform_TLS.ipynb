```python
import pandas as pd
import csv

# Path to your data file
file_path = "tlssample.txt"  # Or .txt if you saved as plain text

# Define column names
columns = [
    "Registration date", "Member Type", "Username", "Fullname", "Line", 
    "Phone", "Contact", "Call Status", "Status", "Amount", 
    "Notes", "TLS Name"
]

# Read the file
df = pd.read_csv(
    file_path,
    sep="\t",                  # tab-separated
    names=columns,             # no header row in file
    header=None,
    engine="python",           # robust with multi-line fields
    quotechar='"',             # honor quotes
    quoting=csv.QUOTE_MINIMAL, # allow newlines inside quoted Notes
    doublequote=True,
    na_values=["null"],
    dtype=str,                 # keep everything as string (incl. '@...')
    keep_default_na=False
)

# optional: normalize line breaks in Notes
df["Notes"] = df["Notes"].str.replace("\r\n", "\n", regex=False).str.strip()

# Move Username column to the first position
# df = df[["Username"] + [col for col in df.columns if col != "Username"]] 

df
# Display the cell of "Notes" from row 5
print(df.at[5, "Notes"])  # Row index is 0-based, so row 5 is index 4 
```
